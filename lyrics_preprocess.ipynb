```python
import pandas as pd
import numpy as np
```

```python
muse_data = pd.read_csv("muse_moreinfo.csv")
muse_data
```

```python
label_file = "id_metadata.csv"
music4all = pd.read_csv(label_file,sep = "	")
music4all
```

```python
label_file = "muse_moreinfo.csv"
music4all_moreinfo = pd.read_csv(label_file)
music4all_moreinfo
```

```python
int_df = pd.merge(muse_data, music4all, how ='inner', on =['spotify_id','spotify_id'])
int_df
```

```python
temp_a = int_df.arousal_tags.values
a_class = [1 if x >= 5 else 0 for x in temp_a]
temp_v = int_df.valence_tags.values
v_class = [1 if x >= 5 else 0 for x in temp_v]
```

```python
from sklearn.utils import class_weight
a_class_weights = class_weight.compute_class_weight(class_weight='balanced',classes = np.unique(a_class), y=a_class)
v_class_weights = class_weight.compute_class_weight(class_weight='balanced',classes = np.unique(v_class),y=v_class)
```

```python
a_sample_weights = np.zeros(shape=(len(a_class),))
for i in range(0,len(a_class)):
    if a_class[i] == 1:
        a_sample_weights[i] = a_class_weights[0]
    else:
        a_sample_weights[i] = a_class_weights[1]
```

```python
lyrics = []
#dataset = dict()
for track_id in int_df.id.values:
    #name = track_id.split('.')
    name = track_id + '.txt'
    path = 'overalap_lyrics' + '/' + name
    f = open(path)
    #text = f.read().strip('\n')
    text = []
    for line in f:
        line = line.strip('\n')
        text.append(line)
    text = ' '.join(text)
    valence_tags = int_df[(int_df['id'] == track_id)]['valence_tags']
    #print(valence_tags)
    if valence_tags.values >= 5:
        valence = 0
        v_sample_weights = v_class_weights[1]
    if valence_tags.values < 5:
        valence = 1
        v_sample_weights = v_class_weights[0]
    arousal_tags = int_df[int_df['id'] == track_id]['arousal_tags']
    if arousal_tags.values >= 5:
        arousal = 0
        a_sample_weights = a_class_weights[1]
    if arousal_tags.values < 5:
        arousal = 1
        a_sample_weights = a_class_weights[0]

    tmp = {'lyrics': text, 'valence': valence, 'arousal':arousal, 'arousal_weight': a_sample_weights, 'valence_weight': v_sample_weights}
    lyrics.append(tmp)
    #lyrics.append(text)
#lyrics = np.asarray(lyrics)
lyrics
```

```python
import json
with open("lyrics.json", "w", encoding="utf8") as f:
        json.dump(lyrics, f)
```
